Company name (Private Inc Limited, Fisher Goods Ltd, Fryer and
Sons, etc.)

1. Build names from components:

- Surname/word list: Fisher, Fryer, Patel, Green, Oak, Beacon, Harbor…

- Industry nouns: Goods, Trading, Logistics, Consulting, Foods, Software, Estates…

- Connectors: & / and / Group / Holdings / Partners

- Legal suffixes: Ltd, Limited, Inc, LLC, GmbH, S.A., BV (choose by “country” if you want)

2. Use templates like:

- {Surname} {Industry} {Suffix} → “Fisher Goods Ltd”

- {Surname} & {Surname} {Suffix} → “Fryer & Sons Limited”

- {Word} {Word} {Suffix} → “Beacon Harbor Inc”

- {Surname} {Connector} {Surname} {Suffix} → “Fisher and Sons Ltd”

```python
import random
import pandas as pd
```

```python
# Generate componests

# First name components (founders, descriptive words)
first_names = [
    "Smith", "Johnson", "Williams", "Brown", "Anderson", "Fisher", "Miller", 
    "Taylor", "Wilson", "Moore", "Clark", "Walker", "Wright", "Robinson",
    "Thompson", "White", "Harris", "Martin", "Garcia", "Martinez", "Lee",
    "Allen", "Young", "King", "Scott", "Green", "Baker", "Adams", "Nelson"
]

# Descriptive/Industry words
descriptive_words = [
    "Global", "Premier", "Elite", "Pacific", "Atlantic", "Northern", "Southern",
    "Eastern", "Western", "Central", "United", "Allied", "National", "Royal",
    "Sterling", "Golden", "Silver", "Diamond", "Apex", "Summit", "Pinnacle",
    "Prime", "First", "Advanced", "Modern", "Classic", "Heritage", "Legacy"
]

# Industry/Business type words
business_words = [
    "Industries", "Enterprises", "Solutions", "Services", "Systems", "Technologies",
    "Consulting", "Partners", "Associates", "Group", "Holdings", "Ventures",
    "Capital", "Investments", "Trading", "Manufacturing", "Logistics", "Dynamics",
    "Innovations", "Resources", "Communications", "Media", "Healthcare", "Properties",
    "Foods", "Goods", "Materials", "Products", "Supplies", "Equipment"
]

# Company suffixes
suffixes = [
    "Inc.", "LLC", "Ltd.", "Co.", "Corp.", "Corporation", "& Co.", 
    "& Sons", "& Associates", "& Partners", "PLC", "LLP", "Group"
]

```

```python
# Generate templates

def generate_company_name():
    """Generate a random company name."""
    pattern = random.choice([1, 2, 3, 4, 5])
    
    if pattern == 1:
        # Fisher Goods Ltd.
        name = f"{random.choice(first_names)} {random.choice(business_words)} {random.choice(suffixes)}"
    elif pattern == 2:
        # Smith & Sons
        name = f"{random.choice(first_names)} {random.choice(['& Sons', '& Co.', '& Associates', '& Partners'])}"
    elif pattern == 3:
        # Global Solutions Inc.
        name = f"{random.choice(descriptive_words)} {random.choice(business_words)} {random.choice(suffixes)}"
    elif pattern == 4:
        # Johnson-Wright Industries LLC
        name = f"{random.choice(first_names)}-{random.choice(first_names)} {random.choice(business_words)}"
    else:
        # Premier Smith Group
        name = f"{random.choice(descriptive_words)} {random.choice(first_names)} {random.choice(suffixes)}"
    
    return name
```

```python
# Company names generation

def generate_company_names(count):
    """Generate multiple unique company names."""
    names = set()
    while len(names) < count:
        names.add(generate_company_name())
    return list(names)
```

```python
company_names = generate_company_names(count = 20000 )
```

```python
# saving
temp_comp_names = pd.DataFrame(company_names, columns=["company"])
temp_comp_names.to_csv("Sub_datasets/companies.csv", index=False)
```
