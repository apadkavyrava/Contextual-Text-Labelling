  Location (Beijing, USA, Boston, Hong Kong, etc.)

```python
# Imports
import pandas as pd
```

```python
# Download found dataset

locations = pd.read_csv('Downloaded_datasets/worldcities.csv')
```

```python
# Remain only needed columns

locations = locations[['city', 'country']]
locations.head(4)
```

```python
# Re-frame in needed format 

locations_long = (
    pd.concat([locations["city"], locations["country"]], axis=0)
      .dropna()
      .reset_index(drop=True)
      .to_frame(name="location")
)

locations_long.head()
```

```python
locations_long.to_csv('Sub_datasets/location.csv')
```
